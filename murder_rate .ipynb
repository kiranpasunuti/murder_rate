```python
data=pd.read_csv('crime_data.csv')
data.head()
```

```python
plt.scatter(data['Murder'],data['Assault'],data['UrbanPop'],data['Rape'])
```

```python
data.corr()
```

```python
x=data.iloc[:,1:]
```

```python
scaler=MinMaxScaler()
scaled_df=scaler.fit_transform(x)
```

```python

scaled_df#labels=scaled_df.fit_predict(x)
```

```python
kmeans=KMeans(n_clusters=2)
kmeans
```

```python
labels=kmeans.fit_predict(x)
labels
```

```python
plt.scatter(scaled_df[:,0],scaled_df[:,1],c=labels)
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=200)
  kmeans.fit_predict(scaled_df)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.grid(True)
```
